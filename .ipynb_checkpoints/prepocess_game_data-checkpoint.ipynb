```python
def next_move_up(game, up):
    return [games_json[game]["moves"][str(i)].split() for i in range(1, up + 1)]

def next_move(i):
    return games_json["1"]["moves"][i].split()
```

```python
for i, game in games_json.items():
    moves = next_move_up(i, len(game["moves"].keys()) - 1)
    board = chess.Board()
    for move in moves:
        board.push_san(move[0])
        if len(move) == 2:
            board.push_san(move[1])
    print(f"Game {i} has been played successfully.")
```

```python
unique_cells = sorted(set("""
r n b q k b n r
p p p p . p p p
. . . . . . . .
. . . . p . . .
. . . . P . . .
. . . . . N . .
P P P P . P P P
R N B Q K B . R"""))

unique_cells.remove("\n")
unique_cells.remove(" ")

unique_figures = np.array(list(set([x.upper() for x in unique_cells[1:]])))

unique_cells, unique_figures
```

```python
fig_dict = {n : i for i, n in enumerate(unique_cells)}
num_dict = {i : n for i, n in enumerate(unique_cells)}

def fig_to_num(f): return fig_dict[f]
def num_to_fig(n): return num_dict[n]
```

```python
def board_to_input(board):
    inp = []
    board = str(board).replace("\n", '').replace(" ", '')
    board.replace(" ", '')
    for s in board:
        inp.append(fig_to_num(s))
    return inp

def input_to_board(inp):
    board = []
    for n in inp:
        board.append(num_to_fig(n))
    return np.array(board).reshape(8, 8)
```

```python
board_pos_dict = {p : n for n, p in enumerate("abcdefgh")}

def board_pos_to_num_pos(b_pos):
#     if len(b_pos) != 2:
#         raise Exception(f"Board position cannot be {b_pos}")
    return np.array([8 - int(b_pos[1]), board_pos_dict[b_pos[0]]])
```

```python
board_pos_to_num_pos("f3")
```

```python
def move_to_output(board, uci_move):
    output = np.empty(10)
    from_pos = board_pos_to_num_pos(uci_move[:2])
    to_pos = board_pos_to_num_pos(uci_move[2:])
    
    # set the figure
    fig_type = board[from_pos[0], from_pos[1]]
    output[:6] = np.zeros(6)
    output[:6][np.where(unique_figures == fig_type.upper())[0]] = 1
    
    # set the pos
    output[6:8] = from_pos / 8
    output[8:10] = to_pos / 8
    return output
```

**Take only games where Carlsen played with white.**

```python
white_games = {}

for k, v in games_json.items():
    if "Carlsen" in v["White"]:
        white_games[k] = v
```

```python
white_games
```

```python
input_data = []
output_data = []

for k, v in white_games.items():
    board = chess.Board()
    for i in v["moves"]:
        if i == "result":
            continue
        moves = v["moves"][i].split()
        inp = board_to_input(board)
        input_data.append(inp)
        move_uci = board.parse_san(moves[0])
        
        output_data.append(move_to_output(input_to_board(inp), str(move_uci)))
        board.push_san(moves[0])
        if len(moves) == 2:
            board.push_san(moves[1])
```

```python
input_data = np.array(input_data)
output_data = np.array(output_data)
output_data
```
